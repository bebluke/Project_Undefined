```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException, ElementClickInterceptedException
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime
import random

def setup_driver():
    chrome_options = Options()
    # chrome_options.add_argument('--headless')
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--lang=zh-TW')
    chrome_options.add_argument('--window-size=1920,1080')
    return webdriver.Chrome(options=chrome_options)

def random_sleep():
    """隨機等待時間，避免被偵測為機器人"""
    time.sleep(random.uniform(1.5, 3.5))

def click_more_reviews(driver):
    """點擊「更多評論」按鈕"""
    try:
        # 尋找並點擊「更多評論」按鈕
        more_buttons = driver.find_elements(By.CSS_SELECTOR, "button.w8nwRe.kyuRq")
        for button in more_buttons:
            try:
                if button.is_displayed() and button.is_enabled():
                    driver.execute_script("arguments[0].click();", button)
                    random_sleep()
            except ElementClickInterceptedException:
                driver.execute_script("arguments[0].scrollIntoView(true);", button)
                random_sleep()
                driver.execute_script("arguments[0].click();", button)
    except Exception as e:
        print(f"點擊更多評論時發生錯誤：{e}")

def scroll_and_load(driver, target_reviews=500, max_attempts=50):
    """
    滾動頁面直到達到目標評論數量或達到最大嘗試次數
    """
    print(f"目標評論數量：{target_reviews}")
    attempts = 0
    last_review_count = 0
    
    while attempts < max_attempts:
        # 獲取當前評論數量
        reviews = driver.find_elements(By.CLASS_NAME, "MyEned")
        current_review_count = len(reviews)
        
        print(f"當前評論數量：{current_review_count}")
        
        # 如果達到目標數量，退出循環
        if current_review_count >= target_reviews:
            print("已達到目標評論數量")
            break
            
        # 如果連續兩次數量相同，嘗試點擊更多評論
        if current_review_count == last_review_count:
            click_more_reviews(driver)
            
        # 滾動到底部
        driver.execute_script(
            "document.querySelector('div.m6QErb.DxyBCb.kA9KIf.dS8AEf').scrollTop = document.querySelector('div.m6QErb.DxyBCb.kA9KIf.dS8AEf').scrollHeight"
        )
        
        # 隨機等待
        random_sleep()
        
        # 更新計數器
        attempts += 1
        last_review_count = current_review_count
        
        if attempts % 5 == 0:
            print(f"已嘗試 {attempts} 次滾動，當前評論數：{current_review_count}")
    
    return current_review_count

def extract_review_info(review_element):
    try:
        # 評論者名稱
        name = review_element.find('div', class_='d4r55').text.strip() if review_element.find('div', class_='d4r55') else "未知"
        
        # 評分
        rating = len(review_element.find_all('img', class_='hCCjke vzX5Ic')) if review_element.find_all('img', class_='hCCjke vzX5Ic') else 0
        
        # 評論時間
        time_element = review_element.find('span', class_='rsqaWe')
        review_time = time_element.text.strip() if time_element else "未知"
        
        # 評論內容
        content_element = review_element.find('span', class_='wiI7pd')
        print('評論',content_element)
        
        # 檢查是否有完整內容按鈕
        full_content_button = review_element.find('button', class_='w8nwRe kyuRq')
        content = "無內容"
        
        if content_element:
            content = content_element.text.strip()
        
        return {
            'reviewer_name': name,
            'rating': rating,
            'review_time': review_time,
            'content': content
        }
    except Exception as e:
        print(f"解析評論時發生錯誤：{e}")
        return None

def scrape_google_maps_reviews(url, target_reviews=500):
    driver = setup_driver()
    reviews_data = []
    
    try:
        print("開始訪問網頁...")
        driver.get(url)
        
        # 等待評論區域加載
        wait = WebDriverWait(driver, 10)
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "MyEned")))
        
        print("開始滾動頁面...")
        total_reviews = scroll_and_load(driver, target_reviews)
        
        # 獲取頁面源碼並解析
        soup = BeautifulSoup(driver.page_source, 'lxml')
        reviews = soup.find_all(class_="MyEned")
        
        print(f"找到 {len(reviews)} 則評論")
        
        # 處理每則評論
        for review in reviews:
            review_info = extract_review_info(review)
            if review_info:
                reviews_data.append(review_info)
        
        # 創建 DataFrame
        df = pd.DataFrame(reviews_data)
        
        # 儲存到 CSV
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        filename = f'google_maps_reviews_{timestamp}.csv'
        df.to_csv(filename, index=False, encoding='utf-8-sig')
        print(f"評論已儲存到 {filename}")
        
        return df
        
    except Exception as e:
        print(f"爬取過程發生錯誤：{e}")
        return None
        
    finally:
        driver.quit()

# 使用範例
if __name__ == "__main__":
    url = "https://www.google.com.tw/maps/place/%E5%8F%B0%E5%8C%97101%E8%B3%BC%E7%89%A9%E4%B8%AD%E5%BF%83/@25.0341222,121.5640212,17z/data=!3m1!5s0x3442abb6d95eb43b:0xbfc3edc9e6aa3050!4m8!3m7!1s0x3442abb6da80a7ad:0xacc4d11dc963103c!8m2!3d25.0341222!4d121.5640212!9m1!1b1!16s%2Fg%2F11fx91ft3n?hl=zh-TW&entry=ttu&g_ep=EgoyMDI1MDEwOC4wIKXMDSoASAFQAw%3D%3D"
    
    # 設定目標評論數量
    target_reviews = 60000
    df = scrape_google_maps_reviews(url, target_reviews)
    
    if df is not None:
        print("\n評論統計：")
        print(f"總評論數：{len(df)}")
        print("\n評分分布：")
        print(df['rating'].value_counts().sort_index())
        
        # 顯示部分評論範例
        print("\n評論範例：")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException, ElementClickInterceptedException
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime
import random

def setup_driver():
    chrome_options = Options()
    # 加入無痕模式設定
    chrome_options.add_argument('--incognito')
    # chrome_options.add_argument('--headless')
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--lang=zh-TW')
    chrome_options.add_argument('--window-size=1920,1080')
    
    # 加入其他可能有用的隱私設定
    chrome_options.add_argument('--disable-blink-features=AutomationControlled')
    chrome_options.add_argument('--disable-extensions')
    chrome_options.add_experimental_option('excludeSwitches', ['enable-automation'])
    chrome_options.add_experimental_option('useAutomationExtension', False)
    
    driver = webdriver.Chrome(options=chrome_options)
    
    # 移除 navigator.webdriver 標記
    driver.execute_cdp_cmd('Page.addScriptToEvaluateOnNewDocument', {
        'source': '''
            Object.defineProperty(navigator, 'webdriver', {
                get: () => undefined
            })
        '''
    })
    
    return driver


def random_sleep():
    """隨機等待時間，避免被偵測為機器人"""
    time.sleep(random.uniform(1.5, 3.5))

def click_more_reviews(driver):
    """點擊「更多評論」按鈕"""
    try:
        more_buttons = driver.find_elements(By.CSS_SELECTOR, "button.w8nwRe.kyuRq")
        for button in more_buttons:
            try:
                if button.is_displayed() and button.is_enabled():
                    driver.execute_script("arguments[0].click();", button)
                    random_sleep()
            except ElementClickInterceptedException:
                driver.execute_script("arguments[0].scrollIntoView(true);", button)
                random_sleep()
                driver.execute_script("arguments[0].click();", button)
    except Exception as e:
        print(f"點擊更多評論時發生錯誤：{e}")

def scroll_and_load(driver, target_reviews=500, max_attempts=50):
    """滾動頁面直到達到目標評論數量或達到最大嘗試次數"""
    print(f"目標評論數量：{target_reviews}")
    attempts = 0
    last_review_count = 0
    
    while attempts < max_attempts:
        reviews = driver.find_elements(By.CLASS_NAME, "MyEned")
        current_review_count = len(reviews)
        
        print(f"當前評論數量：{current_review_count}")
        
        if current_review_count >= target_reviews:
            print("已達到目標評論數量")
            break
            
        if current_review_count == last_review_count:
            click_more_reviews(driver)
            
        driver.execute_script(
            "document.querySelector('div.m6QErb.DxyBCb.kA9KIf.dS8AEf').scrollTop = document.querySelector('div.m6QErb.DxyBCb.kA9KIf.dS8AEf').scrollHeight"
        )
        
        random_sleep()
        
        attempts += 1
        last_review_count = current_review_count
        
        if attempts % 5 == 0:
            print(f"已嘗試 {attempts} 次滾動，當前評論數：{current_review_count}")
    
    return current_review_count

def get_original_content(driver, review_element):
    """點擊顯示原文按鈕並獲取原文內容"""
    try:
        show_original_button = review_element.find_element(By.CSS_SELECTOR, "button[data-review-id]")
        if show_original_button:
            driver.execute_script("arguments[0].click();", show_original_button)
            random_sleep()
            # 等待原文加載
            updated_content = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CLASS_NAME, "original-content"))
            )
            return updated_content.text.strip()
    except Exception as e:
        print(f"獲取原文時發生錯誤：{e}")
    return None

def extract_review_info(review_element):
    try:
        # 評論者名稱
        name = review_element.find('div', class_='d4r55').text.strip() if review_element.find('div', class_='d4r55') else "未知"
        
        # 評分
        rating = len(review_element.find_all('img', class_='hCCjke vzX5Ic')) if review_element.find_all('img', class_='hCCjke vzX5Ic') else 0
        
        # 評論時間
        time_element = review_element.find('span', class_='rsqaWe')
        review_time = time_element.text.strip() if time_element else "未知"
        
        content_translated = "無內容"
        content_original = "無原始內容"
        is_translated = False
        
        content_element = review_element.find('span', class_='wiI7pd')
        
        if content_element:
            # 檢查是否有「顯示原文」按鈕
            show_original_button = content_element.find('button', {'data-review-id': True, 'class': 'w8nwRe kyuRq'})
            
            # 檢查翻譯內容
            translated_content = content_element.find('span', class_='translated-content')
            original_content = content_element.find('span', class_='original-content')
            
            if translated_content and original_content:
                content_translated = translated_content.text.strip()
                content_original = original_content.text.strip()
                is_translated = True
            elif show_original_button:
                content_translated = content_element.text.strip()
                content_original = "[需要點擊顯示原文按鈕]"
                is_translated = True
            else:
                content = content_element.text.strip()
                content_translated = content
                content_original = content
                is_translated = False
        
        return {
            'reviewer_name': name,
            'rating': rating,
            'review_time': review_time,
            'content_translated': content_translated,
            'content_original': content_original,
            'is_translated': is_translated
        }
    except Exception as e:
        print(f"解析評論時發生錯誤：{e}")
        return None

def scrape_google_maps_reviews(url, target_reviews=500):
    driver = setup_driver()
    reviews_data = []
    
    try:
        print("開始訪問網頁...")
        driver.get(url)
        
        wait = WebDriverWait(driver, 10)
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "MyEned")))
        
        print("開始滾動頁面...")
        total_reviews = scroll_and_load(driver, target_reviews)
        
        print("開始處理評論...")
        reviews = driver.find_elements(By.CLASS_NAME, "MyEned")
        
        for i, review in enumerate(reviews, 1):
            print(f"處理第 {i}/{len(reviews)} 則評論")
            
            # 先獲取基本資訊
            review_html = review.get_attribute('outerHTML')
            review_info = extract_review_info(BeautifulSoup(review_html, 'lxml'))
            
            if review_info:
                # 如果發現是翻譯內容，嘗試獲取原文
                if review_info['is_translated']:
                    original_content = get_original_content(driver, review)
                    if original_content:
                        review_info['content_original'] = original_content
                
                reviews_data.append(review_info)
        
        # 創建 DataFrame
        df = pd.DataFrame(reviews_data)
        
        # 儲存到 CSV
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        filename = f'google_maps_reviews_{timestamp}.csv'
        df.to_csv(filename, index=False, encoding='utf-8-sig')
        print(f"評論已儲存到 {filename}")
        
        return df
        
    except Exception as e:
        print(f"爬取過程發生錯誤：{e}")
        return None
        
    finally:
        driver.quit()

# 使用範例
if __name__ == "__main__":
    url = "https://www.google.com.tw/maps/place/%E5%8F%B0%E5%8C%97101%E8%B3%BC%E7%89%A9%E4%B8%AD%E5%BF%83/@25.0335848,121.5536159,15z/data=!3m1!5s0x3442abb6d95eb43b:0xbfc3edc9e6aa3050!4m8!3m7!1s0x3442abb6da80a7ad:0xacc4d11dc963103c!8m2!3d25.0341222!4d121.5640212!9m1!1b1!16s%2Fg%2F11fx91ft3n?hl=zh-TW&entry=ttu&g_ep=EgoyMDI1MDEwOC4wIKXMDSoASAFQAw%3D%3D"
    df = scrape_google_maps_reviews(url, target_reviews=1000)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException, ElementClickInterceptedException
from bs4 import BeautifulSoup
import time
import pandas as pd
from datetime import datetime
import random

def setup_driver():
    chrome_options = Options()
    # 加入無痕模式設定
    chrome_options.add_argument('--incognito')
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--lang=zh-TW')
    chrome_options.add_argument('--window-size=1920,1080')
    
    # 加入其他可能有用的隱私設定
    chrome_options.add_argument('--disable-blink-features=AutomationControlled')
    chrome_options.add_argument('--disable-extensions')
    chrome_options.add_experimental_option('excludeSwitches', ['enable-automation'])
    chrome_options.add_experimental_option('useAutomationExtension', False)
    
    driver = webdriver.Chrome(options=chrome_options)
    
    # 移除 navigator.webdriver 標記
    driver.execute_cdp_cmd('Page.addScriptToEvaluateOnNewDocument', {
        'source': '''
            Object.defineProperty(navigator, 'webdriver', {
                get: () => undefined
            })
        '''
    })
    
    return driver

def random_sleep():
    """隨機等待時間，避免被偵測為機器人"""
    time.sleep(random.uniform(2, 4))

def click_more_reviews(driver):
    """點擊「更多評論」按鈕"""
    max_retries = 3
    for retry in range(max_retries):
        try:
            # 等待按鈕出現
            WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CSS_SELECTOR, "button.w8nwRe.kyuRq"))
            )
            
            more_buttons = driver.find_elements(By.CSS_SELECTOR, "button.w8nwRe.kyuRq")
            for button in more_buttons:
                try:
                    if button.is_displayed() and button.is_enabled():
                        # 捲動到按鈕位置
                        driver.execute_script("arguments[0].scrollIntoView(true);", button)
                        random_sleep()
                        # 點擊按鈕
                        driver.execute_script("arguments[0].click();", button)
                        random_sleep()
                        return True
                except Exception as e:
                    print(f"點擊按鈕時發生錯誤：{e}")
                    continue
        except Exception as e:
            print(f"重試 {retry + 1}/{max_retries}: {e}")
            random_sleep()
    return False

def scroll_and_load(driver, target_reviews=1000, max_attempts=100):
    """滾動頁面直到達到目標評論數量或達到最大嘗試次數"""
    print(f"目標評論數量：{target_reviews}")
    attempts = 0
    last_review_count = 0
    no_change_count = 0
    
    while attempts < max_attempts:
        reviews = driver.find_elements(By.CLASS_NAME, "MyEned")
        current_review_count = len(reviews)
        
        print(f"當前評論數量：{current_review_count}")
        
        if current_review_count >= target_reviews:
            print("已達到目標評論數量")
            break
            
        if current_review_count == last_review_count:
            no_change_count += 1
            if no_change_count >= 5:  # 如果連續5次都沒有新評論
                # 嘗試點擊「更多評論」按鈕
                if click_more_reviews(driver):
                    no_change_count = 0  # 重置計數器
                else:
                    # 如果點擊失敗且沒有新評論，可能已經到底
                    print("似乎已經到達所有可用評論的底部")
                    break
        else:
            no_change_count = 0
        
        # 執行滾動
        for _ in range(3):  # 每次嘗試多次小幅度滾動
            driver.execute_script(
                "document.querySelector('div.m6QErb.DxyBCb.kA9KIf.dS8AEf').scrollTop += 300"
            )
            time.sleep(0.5)
        
        random_sleep()
        
        attempts += 1
        last_review_count = current_review_count
        
        if attempts % 5 == 0:
            print(f"已嘗試 {attempts} 次滾動，當前評論數：{current_review_count}")
            time.sleep(2)
    
    return current_review_count
def get_original_content(driver, review_element):
    """點擊顯示原文按鈕並獲取原文內容"""
    try:
        show_original_button = review_element.find_element(By.CSS_SELECTOR, "button[data-review-id]")
        if show_original_button:
            driver.execute_script("arguments[0].click();", show_original_button)
            random_sleep()
            # 等待原文加載
            updated_content = WebDriverWait(driver, 5).until(
                EC.presence_of_element_located((By.CLASS_NAME, "original-content"))
            )
            return updated_content.text.strip()
    except Exception as e:
        print(f"獲取原文時發生錯誤：{e}")
    return None

def extract_review_info(review_element):
    try:
        # 評論者名稱
        name = review_element.find('div', class_='d4r55').text.strip() if review_element.find('div', class_='d4r55') else "未知"
        
        # 評分
        rating = len(review_element.find_all('img', class_='hCCjke vzX5Ic')) if review_element.find_all('img', class_='hCCjke vzX5Ic') else 0
        
        # 評論時間
        time_element = review_element.find('span', class_='rsqaWe')
        review_time = time_element.text.strip() if time_element else "未知"
        
        content_translated = "無內容"
        content_original = "無原始內容"
        is_translated = False
        
        content_element = review_element.find('span', class_='wiI7pd')
        
        if content_element:
            # 檢查是否有「顯示原文」按鈕
            show_original_button = content_element.find('button', {'data-review-id': True, 'class': 'w8nwRe kyuRq'})
            
            # 檢查翻譯內容
            translated_content = content_element.find('span', class_='translated-content')
            original_content = content_element.find('span', class_='original-content')
            
            if translated_content and original_content:
                content_translated = translated_content.text.strip()
                content_original = original_content.text.strip()
                is_translated = True
            elif show_original_button:
                content_translated = content_element.text.strip()
                content_original = "[需要點擊顯示原文按鈕]"
                is_translated = True
            else:
                content = content_element.text.strip()
                content_translated = content
                content_original = content
                is_translated = False
        
        return {
            'reviewer_name': name,
            'rating': rating,
            'review_time': review_time,
            'content_translated': content_translated,
            'content_original': content_original,
            'is_translated': is_translated
        }
    except Exception as e:
        print(f"解析評論時發生錯誤：{e}")
        return None

def scrape_google_maps_reviews(url, target_reviews=500):
    driver = setup_driver()
    reviews_data = []
    
    try:
        print("開始訪問網頁...")
        driver.get(url)
        
        wait = WebDriverWait(driver, 10)
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, "MyEned")))
        
        print("開始滾動頁面...")
        total_reviews = scroll_and_load(driver, target_reviews)
        
        print("開始處理評論...")
        reviews = driver.find_elements(By.CLASS_NAME, "MyEned")
        
        for i, review in enumerate(reviews, 1):
            print(f"處理第 {i}/{len(reviews)} 則評論")
            
            # 先獲取基本資訊
            review_html = review.get_attribute('outerHTML')
            review_info = extract_review_info(BeautifulSoup(review_html, 'lxml'))
            
            if review_info:
                # 如果發現是翻譯內容，嘗試獲取原文
                if review_info['is_translated']:
                    original_content = get_original_content(driver, review)
                    if original_content:
                        review_info['content_original'] = original_content
                
                reviews_data.append(review_info)
        
        # 創建 DataFrame
        df = pd.DataFrame(reviews_data)
        
        # 儲存到 CSV
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        filename = f'google_maps_reviews_{timestamp}.csv'
        df.to_csv(filename, index=False, encoding='utf-8-sig')
        print(f"評論已儲存到 {filename}")
        
        return df
        
    except Exception as e:
        print(f"爬取過程發生錯誤：{e}")
        return None
        
    finally:
        driver.quit()

# 使用範例
if __name__ == "__main__":
    url = "https://www.google.com.tw/maps/place/%E5%9C%8B%E7%AB%8B%E6%95%85%E5%AE%AE%E5%8D%9A%E7%89%A9%E9%99%A2/@25.102355,121.5381928,15z/data=!3m1!5s0x3442ac3b2ddb9a43:0x7a84c798191dd2cf!4m8!3m7!1s0x3442ac3acd404a7d:0x5d6d7018397a09c1!8m2!3d25.1023554!4d121.5484925!9m1!1b1!16zL20vMGhod2w?hl=zh-TW&entry=ttu&g_ep=EgoyMDI1MDEwOC4wIKXMDSoASAFQAw%3D%3D"
    df = scrape_google_maps_reviews(url, target_reviews=1000)
```
